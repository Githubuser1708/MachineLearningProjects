```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
data = pd.read_csv("SNAP.csv")
print(data.head())
print(data.tail())
```

```python
data.tail(20)
```

```python
data = data.drop("Adj Close",axis=1)
```

```python
data
```

```python
data.info()
```

```python
data.shape
```

```python
from autots import AutoTS

```

```python
# To ignore the warnings that may appear
import warnings
warnings.filterwarnings("ignore")
```

```python
model = AutoTS(
    forecast_length=5,
    frequency='infer',
    ensemble='simple',
    max_generations=8,
    num_validations=2,
    no_negatives=True
)
```

```python
model = model.fit(data, date_col='Date', value_col='Close', id_col=None)

# Print the description of the best model
print(model)
```

```python
prediction = model.predict()
forecast = prediction.forecast
print(f"the forecast of SNAP in the next 5 days is: {forecast}")
```
